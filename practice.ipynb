```python
# counting of vowels in the 
vowels=['a','e','i','o','u']
word="vishwa"
count=0
for character in word:
    if character in vowels:
        count+=1
print('number of vowels ',count)
```

```python
vowels=['a','e','i','o','u']
word="vishwanthan"
count=0
for character in word :
    if character not in vowels:
        count+=1
print(count)
```

```python
import numpy as np
```

```python
x=np.array([1, 2, 3, 4, 5, 6, 7, 8])#studying hour
y=np.array([45, 55, 60, 65, 70, 75, 82, 88])#marks
a=np.sum(y)
a
```

```python
n=len(x)
m = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
b = (np.sum(y) - m * np.sum(x)) / n
print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
```

```python
hours = 9
predicted_score = m * hours + b
print(f"Predicted score for {hours} hours: {predicted_score:.1f}")
```

```python
predictions = m * x + b
mse = np.mean((y - predictions)**2)
print(f"MSE: {mse:.2f}")
```
